# Data Preparation and Quality Assurance
The first step in our analysis involved importing the World Bank Excel file, which contains comprehensive economic indicators. We calculated the population in millions by utilizing the "GDP" and "GDP per Capita" values to create a new column titled "Population (M)." Basic profiling and quality assurance checks were conducted to ensure the integrity and completeness of the data. Finally, we filtered the dataset to focus on the year 2014 and merged it with the HDI CSV file based on the "Country Code," setting the stage for deeper analysis and visualization.

```python
import pandas as pd
import numpy as np
world_bank = pd.read_excel("C:/code/WorldBank.xlsx").rename({"Country Name":"Country"},axis=1)
world_bank["Population (M)"]=(world_bank["GDP (USD)"]/world_bank["GDP per capita (USD)"])/1000000
world_bank.head()

```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi=pd.read_csv("C:/code/HDI.csv")
wb_hdi_2014=world_bank.query("Year==2014").merge(hdi[["iso3","hdi_2014"]],how="left",left_on="Country Code",right_on="iso3")
wb_hdi_2014.head()
```

```python
wb_hdi_2014.describe()
```

# Data Transformation for Visualization
In this objective, we focused on reshaping the prepared data into structured tables suitable for visualization.

GDP Pivot Table: We created a "gdp_pivot" table where years are represented as rows and regions as columns. The values in this table reflect the sum of GDP for each region across the specified years, allowing for clear comparisons of economic performance over time.

Population Pivot Table: Similarly, we developed a "pop_pivot" table that organizes years as rows and regions as columns, with the sum of the population as the values. This transformation highlights demographic trends and regional population distributions over time.

Average HDI by Region: Finally, we calculated the average Human Development Index (HDI) for each region to create the "wb_hdi_by_region" table. This table is sorted from highest to lowest average HDI, providing a clear view of regional disparities in human development.

These structured tables will serve as foundational datasets for our subsequent visualizations, enabling us to effectively communicate key insights about global economic trends.

```python
gdp_pivot=world_bank.pivot_table(index="Year",columns="Region",values="GDP (USD)",aggfunc="sum")
gdp_pivot.head()
```

```python
pop_pivot=world_bank.pivot_table(index="Year",columns="Region",values="Population (M)",aggfunc="sum")
pop_pivot.head()
```

```python
wb_hdi_by_region=wb_hdi_2014.groupby("Region").agg({"hdi_2014":"mean"}).sort_values("hdi_2014",ascending=False)
wb_hdi_by_region.head()
```

# Data Visualization
In this objective, we created five compelling visualizations to illustrate key trends, relationships, and comparisons across regions:

Stacked Area Chart of GDP Growth: This chart visualizes the growth of GDP over time, with each stack representing a different geographic region. The visualization highlights how economic growth trajectories differ among regions, allowing for quick identification of leaders and laggards in economic development.

Stacked Area Chart of Population Growth: Similar to the GDP chart, this stacked area chart depicts population growth over time by region. It provides insights into demographic changes, showing which regions are experiencing significant population increases and which are stabilizing or declining.

Bubble Chart of Life Expectancy vs. GDP Per Capita: In this chart, life expectancy is plotted on the x-axis and GDP per capita on the y-axis (using a log scale). The size of each bubble represents population size, and colors indicate different regions. This visualization reveals the relationship between economic prosperity and health outcomes, highlighting disparities in life expectancy across income levels.

Bar Chart of Average HDI by Region: This bar chart displays the average HDI for each region, with colors matching those used in previous charts. It effectively showcases regional differences in human development, emphasizing areas of both high and low HDI.

Scatterplot of Power Consumption vs. GDP Per Capita: The scatterplot presents power consumption on the x-axis and GDP per capita on the y-axis, with HDI represented as a color scale for the data points. By excluding outliers like Iceland, the visualization illustrates the correlation between energy consumption and economic wealth, shedding light on trends in resource use among different countries.

These visualizations collectively provide a comprehensive overview of the global economic landscape, facilitating a deeper understanding of the intricate relationships between economic indicators and human development across regions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig , ax=plt.subplots()
ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index
)
ax.legend(loc="upper left")
ax.set_title("GDP has Grown Exponentially Over Time")
ax.set_ylabel("GDP  (Trillions)")
```

```python

fig , ax=plt.subplots()
ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index
)
ax.legend(loc="upper left")
ax.set_title("Population has surged from 2 Billion to 7.5 Billion")
ax.set_ylabel("Population (Millions)")
```

```python
import matplotlib.ticker as mticker
fig,ax=plt.subplots()

minsize= min(wb_hdi_2014["Population (M)"])
maxsize= max(wb_hdi_2014["Population (M)"])
sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize,maxsize),
    hue="Region"
).set(title="Life Expectancy increases as Countries gets richer",yscale="log")
ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
wb_hdi_by_region.plot.bar(title="HDI by Region",ylabel="Human Development Index (HDI)" )
```

```python
sns.scatterplot(
    data=wb_hdi_2014.query("Country !='Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r",
    legend=True
).set(title="Electricity drives Development")
```

# Building the Final Report
In this objective, we structured the five visualizations into a cohesive one-page report to effectively communicate the state of the global economy.

Assembled Visualizations: The five charts were arranged systematically to tell a clear story about global economic trends and human development:

Stacked Area Chart of GDP Growth
Stacked Area Chart of Population Growth
Bubble Chart of Life Expectancy vs. GDP Per Capita
Bar Chart of Average HDI by Region
Scatterplot of Power Consumption vs. GDP Per Capita
Overall Title: A compelling title was added to encapsulate the theme of the report: "Key Insights into Global Economic Trends and Human Development."

Descriptive Context: Each visualization was accompanied by brief descriptions to provide context and enhance understanding. This text explains the significance of each chart, highlighting key insights and trends observed in the data.

By assembling these elements into a single cohesive report, we created a powerful tool for analyzing and communicating the complexities of the global economy, facilitating a deeper understanding of the interconnections between economic performance and human development across regions.

```python

import matplotlib.gridspec as gridspec
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Arial'


# Create a figure with a grid specification
fig = plt.figure(figsize=(8, 12))
gs = gridspec.GridSpec(nrows=12, ncols=6)

stack1_list = ['#66C2A5','#fC8D62','#8DA8CB','#E78AC3','#A6D854','#FFD92F','#838383'] 
stack2_list = ['#66C2A5', '#A6D854', '#838383', '#fC8D62', '#E78AC3', '#8DA8CB', '#FFD92F'] 
bar_list= ['#8DA0CB','#fC8D62','#E78AC3','#FFD92F','#66C2A5','#A6D854','#B3B3b3'] 
bubble_list = ['#A6D854','#fC8D62','#FFD92F','#66C2A5','#838383','#E78AC3','#8DABCB']
fig.suptitle(
    "Report: Global Economic Development Report from 1960-2018",
    fontsize=16,
    fontweight="bold",
    x=.65,
    y=.93
)


# GDP Stackplot
ax1 = fig.add_subplot(gs[0:4, 0:3])
ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_list
)

ax1.set_title("GDP has Grown Exponentially Over Time")
ax1.set_ylabel("GDP (Trillions)")
fig.text(.92,.62,
"""

Economic Expansion in the 20th and early
21st Centuries has been enormous and has 
outpaced population growth.In 1960, Europe 
and North America produced 15% of GDP on 
20% of Population.

By 2018, East Asia had surpassed both 
in economic output largely due to the 
rise of Japan, South Korea and China.

In the coming decades, economists expect 
South Asia and Sub-Saharan Africa to 
lead the world in global growth.
"""
,fontsize=8)


# Population Stackplot
ax2 = fig.add_subplot(gs[0:4, 3:6])
ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_list
)

ax2.set_title("Population has surged from 2 Billion to 7.5 Billion")
ax2.set_ylabel("Population (Millions)")
ax2.legend(bbox_to_anchor=(1.75,1), fontsize=8)
# Life Expectancy vs GDP per Capita Scatter Plot
ax3 = fig.add_subplot(gs[4:8, :])
minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])
sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize, maxsize),
    hue="Region",
    ax=ax3,
    palette=bubble_list
).set(title="Life Expectancy Increases as Countries Get Richer", yscale="log")
ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())
handles, labels=ax3.get_legend_handles_labels()
entries_to_skip= len(wb_hdi_2014["Region"].unique())+1
for h in handles[1:]:
     if hasattr(h, 'get_sizes'):
        sizes=[s/1.5 for s in h.get_sizes()]
        h.set_sizes(sizes)
ax3.legend(handles[entries_to_skip:],
           labels[entries_to_skip:],
           bbox_to_anchor=(.8,.9),
           borderaxespad=0,
           frameon=False,
           ncols=6,
           fontsize=8
          )
fig.text(.92,.45,
"""

he wealthy nations of the world enjoy 
high GDP per capita as well as well 
as long life spans, but make up a 
relatively small share of the global 
population.
If growth can continue in the developing 
world, humanity will be vastly wealthier, 
healthier, and (hopefully) happier.
"""
,fontsize=8)           

# HDI by Region Bar Plot
ax4 = fig.add_subplot(gs[8:12, 0:3])
ax4=wb_hdi_by_region.plot.bar(ax=ax4,title="HDI by Region", ylabel="Human Development Index (HDI)",color=bar_list)


# Electricity Consumption vs GDP per Capita Scatter Plot
ax5 = fig.add_subplot(gs[8:12, 3:6])
sns.scatterplot(
    data=wb_hdi_2014.query("Country != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r",
    ax=ax5,
    legend=True
).set(title="Electricity Drives Development")
fig.text(.92,.16,
"""


HDI, short for Human Development Index, 
attempts to measure the overall 
standard of living in countries. 
Life Expectancy, GDP per Capita, 
and Educational attainment are the 
factors that are considered.
Economic growth in the developing 
world in the 21st century should help 
other regions catch up to North America
and Europe.
One factor not in HDI, but instrumental in
a country's development, is electricty 
consumption. Electricity unlocks massive 
improvements in productivity. Developing 
nations should continue to invest in energy 
production to ensure growth.
"""
,fontsize=8)         

# Adjust layout and show the figure
fig.subplots_adjust(wspace=4,hspace=4)
```

# Correlation
What is the correlation between GDP per capita and HDI in 2014?

```python
data_2014 = wb_hdi_2014[wb_hdi_2014['Year'] == 2014][['GDP per capita (USD)', 'hdi_2014']]
data_2014 = data_2014.dropna()
correlation = data_2014.corr().iloc[0, 1]

print(f"The correlation between GDP per capita and HDI in 2014 is: {correlation:.4f}")
```
